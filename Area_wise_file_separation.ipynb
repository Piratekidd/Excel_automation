```python
import pandas as pd
import os  # Import os module for handling file paths
from datetime import datetime

# Replace 'file_path.xlsx' with the path to your Excel file
df = pd.read_excel("F:\Daily Reports\center_size.xlsx")

# Filter the DataFrame based on the 'Active Center' column
filtered_df = df[(df['Active Center'] >= 2) | ((df['Active Center'] < 2) & (df['Active center size'] >= 4))]

# Exclude rows with specific values in the 'app_status' column
filtered_df = filtered_df[~filtered_df['app_status'].isin(['REJECTED', 'DISBURSED', 'CB_VERIFICATION_FAILED','FOIR_REJECTED','CREDIT_SUBMIT_REJECTED','CLOSED','CBC_ERROR'])]

# Group the filtered DataFrame by the 'Area Name' column
grouped = filtered_df.groupby('Area name')

# Specify the directory where you want to save the Excel files
directory = "F:\Area"

# Create the 'Area' folder if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)


# Get today's date
today_date = datetime.today().strftime('%Y-%m-%d')

# Iterate over the groups and save each group as a separate Excel file
for area, data in grouped:
    # Create the file path for each Excel file inside the 'Area' folder
    file_count = len(data)
    file_name = os.path.join(directory, f'{area}_{file_count}_{today_date}.xlsx')
    with pd.ExcelWriter(file_name) as writer:
        data.to_excel(writer, index=False)
```

```python
import pandas as pd
import os
from datetime import datetime

# Replace 'file_path.xlsx' with the path to your Excel file
df = pd.read_excel("F:\Daily Reports\center_size.xlsx")

# Filter the DataFrame based on the 'Active Center' column
filtered_df = df[(df['Active Center'] >= 2) | ((df['Active Center'] < 2) & (df['center size'] > 3))]

# Exclude rows with specific values in the 'app_status' column
filtered_df = filtered_df[~filtered_df['app_status'].isin(['REJECTED', 'DISBURSED', 'CB_VERIFICATION_FAILED'])]

# Group the filtered DataFrame by the 'Area Name' column
grouped = filtered_df.groupby('Area name')

# Specify the directory where you want to save the Excel files
directory = "F:\Area"

# Create the 'Area' folder if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Get today's date
today_date = datetime.today().strftime('%Y-%m-%d')

# Iterate over the groups and save each group as a separate Excel file
for area, data in grouped:
    # Create the file path for each Excel file inside the 'Area' folder
    file_count = len(data)
    file_name = os.path.join(directory, f'{area}_{file_count}records_{today_date}.xlsx')
    
    # Create a Pandas Excel writer for each area-wise file using XlsxWriter as the engine
    with pd.ExcelWriter(file_name, engine='xlsxwriter') as writer:
        # Write the data to the first sheet
        data.to_excel(writer, sheet_name='Data', index=False)
        
        # Create pivot table in another sheet
        pivot_table_df = data.pivot_table(index='app_status', columns='Center Type', values='Count of loan_id', aggfunc='count')
        pivot_table_df.to_excel(writer, sheet_name='Pivot_Table')
```
